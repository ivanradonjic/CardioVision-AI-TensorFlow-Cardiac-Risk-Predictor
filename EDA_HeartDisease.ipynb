**Exploratory Data Analysis**


**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Import**

```python
df = pd.read_csv("heart_statlog_cleveland_hungary_final.csv")
```

```python
df.info()
```

**Remove Duplicates**

```python
df.drop_duplicates(inplace = True)
df.reset_index(inplace = True, drop = True)
df.info()
```

**Exploratory Data Analysis**

```python
#Split the dataset into features and targets
X = df.drop('target', axis = 1)
y = df['target']
```

```python
#Plot Distribution of age

#Extract Data
age_counts = X['age'].value_counts()

# Plotting the histogram with seaborn
plt.figure(figsize=(10, 6))
sns.histplot(X['age'], bins=20, kde=True, color='blue')

# Adding labels and title
plt.ylabel('Count')
plt.xlabel('Age')
plt.title('Distribution of Age')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
# Plot Distribution of Resting Blood Pressure

plt.figure(figsize=(10, 6))
sns.histplot(X['resting bp s'], bins=20, kde=True, color='blue')

# Adding labels and title
plt.ylabel('Count')
plt.xlabel('Resting Blood Pressure')
plt.title('Distribution of Resting Blood Pressure')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
#Plot Distribution of Cholesterol Levels

plt.figure(figsize=(10, 6))
sns.histplot(X['cholesterol'], bins=20, kde=True, color='blue')

# Adding labels and title
plt.ylabel('Count')
plt.xlabel('Cholesterol Levels')
plt.title('Distribution of Cholesterol Levels')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
#Plot Distribution of Max Heart Rate Levels

plt.figure(figsize=(10, 6))
sns.histplot(X['max heart rate'], bins=20, kde=True, color='blue')

# Adding labels and title
plt.ylabel('Count')
plt.xlabel('Max Heart Rate Levels')
plt.title('Distribution of Max Heart Rate Levels')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
#Plot Distribution of Oldpeak

oldpeak_counts = X['oldpeak'].value_counts()

# Plotting the histogram with seaborn
plt.figure(figsize=(10, 6))
sns.histplot(X['oldpeak'], bins=20, kde=True, color='blue', edgecolor='black')

# Adding labels and title
plt.ylabel('Count')
plt.xlabel('oldpeak')
plt.title('Distribution of Oldpeak')
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
#Plot Distribution of Chest Pain Type

chest_pain_type = X['chest pain type'].value_counts().sort_index(ascending=True)
chest_pain_type.plot(kind = 'bar', color = ['blue', 'goldenrod', 'orange', 'red'], edgecolor='black')
chest_pain_type.index = ['1', '2', '3', '4']
plt.ylabel('Count')
plt.xlabel('Chest Pain Type')
plt.xticks(rotation=0)
plt.title('Distribution of Chest Pain Type')
plt.show()
```

```python
#Plot Distribution of Males and Females

sex_counts = X['sex'].value_counts()
print(sex_counts)

sex_counts.plot(kind = 'bar', color = ['blue', 'orange'], edgecolor='black')
plt.ylabel('Count')
plt.xlabel('Sex (1 = Male, 0 = Female)')
plt.title('Distribution of Males and Females')
plt.xticks(rotation=0)
plt.show()
```

```python
#Plot Distribution of Fasting Blood Sugar Levels

fbs_counts = X['fasting blood sugar'].value_counts()
print(fbs_counts)

fbs_counts.plot(kind = 'bar', color = ['blue', 'orange'], edgecolor='black')
plt.ylabel('Count')
plt.xlabel('Fasting Blood Sugar > 120 mg/dl (1 = true; 0 = false)')
plt.title('Distribution of Fasting Blood Sugar Levels')
plt.xticks(rotation=0)
plt.show()
```

```python
#Plot Distribution of Exersice Incuded Angina

exercise_angina_counts = X['exercise angina'].value_counts()
print(exercise_angina_counts)

exercise_angina_counts.plot(kind = 'bar', color = ['blue', 'orange'], edgecolor='black')
plt.ylabel('Count')
plt.xlabel('Exersice Incuded Angina? (1 = yes; 0 = no)')
plt.title('Distribution of Exersice Incuded Angina')
plt.xticks(rotation=0)
plt.show()
```

```python
#Plot Distribution of Resting ECG

ecg_counts = X['resting ecg'].value_counts()
print(ecg_counts)

ecg_counts.plot(kind = 'bar', color = ['blue', 'orange', 'red'], edgecolor='black')
plt.ylabel('Count')
plt.xlabel('Resting ECG (0 - Normal, 1 - Abnormal, 2 - Definite Hypertophy)')
plt.title('Distribution of Resting ECG')
plt.xticks(rotation=0)
plt.show()
```

```python
#Plot Distribution of ST Slope

STslope_counts = X['ST slope'].value_counts()
print(STslope_counts)

ecg_counts.plot(kind = 'bar', color = ['blue', 'orange', 'red'], edgecolor='black')
plt.ylabel('Count')
plt.xlabel('ST Slope (1 = Upsloping, 2 - Flat, 3 - Down Sloping)')
plt.title('Distribution of ST Slope')
plt.xticks(rotation=0)
plt.show()
```

```python
#Plot Distribution of Heart Disease
heart_disease_counts = y.value_counts()
print(heart_disease_counts)

heart_disease_counts.plot(kind = 'bar', color = ['blue', 'orange'], edgecolor='black')
plt.ylabel('Count')
plt.xlabel('Patients with Heart Disease (1 - Heart Disease, 0 - No Heart Disease)')
plt.title('Distribution of Heart Disease')
plt.xticks(rotation=0)
plt.show()
```
